```python
### 시각화 라이브러리 불러들이기
import seaborn as sns

import matplotlib.pyplot as plt
```

```python
### 엠스콤 데이터프레임으로 읽어들이기
ans = sns.load_dataset("anscombe")
ans
```

```python
### 데이터 정보 확인 : 결측치 확인 가능
ans.info()
```

```python
### 기초통계 데이터 확인하기
ans.describe()
```

```python
### dataset 컬럼별로 그룹지어 주세요
ans_group = ans.groupby(['dataset'])
# head()로 그룹된 항목들을 확인할 수 있음 (2) -> 그룹의 갯수만큼 확인 가능
ans_group.head(2)
```

```python
### 그룹화한 데이터 기초통계하기
ans_group.describe()
```

```python
### I그룹 데이터 추출하기
# data1 = ans[ans['dataset'] == "I"]
filter = ans['dataset']=="I"
data1 = ans[filter]

data1 = ans[ans['dataset'] == "I"]
data1
```

```python
plt.plot(data1["x"], data1["y"], "o")
```

```python
data2 = ans[ans['dataset'] == "II"]
data3 = ans[ans['dataset'] == "III"]
data4 = ans[ans['dataset'] == "IV"]
```

```python
plt.plot(data2["x"], data2['y'], "o")
```

```python
plt.plot(data3["x"], data3['y'], "o")
# 이상치 데이터를 가짐
```

```python
plt.plot(data4["x"], data4['y'], "o")
# 이상치 데이터 가짐
```

```python
### 하나의 그래프에 4개의 각 그래프 표현하기

# 하나의 그래프 객체 선언
fig = plt.figure()

### 4개의 그래프가 들어갈 공간 만들기
# - add_subplot(행, 열, 위치)
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
ax4 = fig.add_subplot(2, 2, 4)

### 각 공간 그래프에 제목 달기
ax1.set_title("data1")
ax2.set_title("data2")
ax3.set_title("data3")
ax4.set_title("data4")

# ### 그래프 겹치지 않게 정렬하기
# fig.tight_layout()
# tight_layout()은 마지막에 한번만 넣으면 된다.

### 그래프에 데이터 넣어서 그리기
ax1.plot(data1["x"], data1["y"], "o", c="b")
ax2.plot(data2["x"], data2["y"], "o", c="r")
ax3.plot(data3["x"], data3["y"], "o", c="g")
ax4.plot(data4["x"], data4["y"], "o", c="y")

### 그래프 전체 제목 넣기
fig.suptitle("Anscombe Data")

### 그래프 겹치지 않게 정렬하기
fig.tight_layout()
```

```python
### 시각화 그래프 저장시키기
# - ./ : 주피터는 현재 로컬 위치를 기준으로 하기 때문에
fig.savefig("./fig.png")
```
